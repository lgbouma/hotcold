```python
import lightkurve as lk
import numpy as np, matplotlib.pyplot as plt
%matplotlib inline
```

```python
r = lk.search_lightcurve("TIC 141146667")
search_result = lk.search_targetpixelfile("TIC 141146667")
print(r)
print(search_result)
```

```python
lc0 = r[2].download()
tpf0 = search_result[2].download()
```

```python
time = np.array(lc0.time.value)
flux = np.array(lc0.pdcsap_flux.value)
plt.figure(figsize=(15,7))
plt.scatter(time, flux, s=1);
```

```python
tpf0.interact_sky()
```

```python
transform_func = lambda lc: lc.remove_nans().normalize()  
tpf0.interact(transform_func=transform_func)
```
